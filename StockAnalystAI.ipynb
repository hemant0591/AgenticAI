```python
from agno.agent import Agent
from agno.models.google import Gemini
#from agno.models.openai import OpenAIChat
from agno.tools.reasoning import ReasoningTools
from agno.tools.yfinance import YFinanceTools
```

```python
from google.colab import userdata
import os

os.environ['GOOGLE_API_KEY'] = userdata.get('Gemini-key')
```

```python
agent = Agent(
  model = Gemini(id="gemini-2.0-flash"),
  #model = OpenAIChat(id="gpt-4o"),
  tools = [
      ReasoningTools(add_instructions=True),
      YFinanceTools(),
  ],
  instructions = [
      "Use table to display data",
      "Only output reports, no other text",
  ],
  markdown = True
)
```

```python
agent.print_response(input="Write a report on NVDA", stream=True, show_full_reasoning=True, stream_intermediate_steps=True)
```
